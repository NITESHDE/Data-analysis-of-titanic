```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_data = sns.load_dataset('titanic')
```

```python
# Display the first few rows of the DataFrame
titanic_data.head()
```

```python
# Display basic statistical information about the DataFrame
titanic_data.describe()
```

```python
# Display the data types of each column in the DataFrame
titanic_data.dtypes
```

```python
# Select a specific column from the DataFrame
titanic_data['class']
```

```python
# Filter the DataFrame based on a condition
titanic_data[titanic_data['alive'] == 'yes']
```

```python
# Create a line plot
plt.plot(titanic_data['age'], titanic_data['fare'], label ='Data points')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
# Create a scatter plot
plt.scatter(titanic_data['age'], titanic_data['fare'],label = 'Data points')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

```python
# Create a histogram
plt.hist(titanic_data['age'])
plt.xlabel('Age')
plt.ylabel('fare')
plt.show()
```
